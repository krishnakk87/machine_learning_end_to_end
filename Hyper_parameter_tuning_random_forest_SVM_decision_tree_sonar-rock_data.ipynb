```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sonar-all-data.csv",header=None)
```

```python
df.info()
```

```python
for i in range(0,208):
    if(df.iloc[i,-1]=='R'):
        df.iloc[i,-1]=0
    else:
        df.iloc[i,-1]=1
```

```python
df.head(4)
```

```python
df.corr()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
y=y.astype('int')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

classifier_svc = SVC(kernel = 'rbf', random_state = 0)
classifier_tree = DecisionTreeClassifier(random_state = 0)
classifier_rf = RandomForestClassifier(random_state = 0)
```

```python
accuracies_svc = cross_val_score(estimator = classifier_svc, X = X_train, y = y_train, cv = 10)
accuracies_tree = cross_val_score(estimator = classifier_tree, X = X_train, y = y_train, cv = 10)
accuracies_rf = cross_val_score(estimator = classifier_rf, X = X_train, y = y_train, cv = 10)
```

```python
print('accuracies_svc cross validation: ',np.mean(accuracies_svc))
print('accuracies_tree cross validation: ',np.mean(accuracies_tree))
print('accuracies_rf cross validation: ',np.mean(accuracies_rf))
```

##### random forest has highest cross validation accuracy on trainig set, further we will fine tune random forest hyper parameters

```python
from sklearn.model_selection import GridSearchCV

# Number of trees in random forest
n_estimators = [100,150,200,250]
# Number of features to consider at every split
max_features = ['sqrt', 'log2',None]
# Maximum number of levels in tree
max_depth = [6,10,14,18,22,26,30]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 4,6]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4,6,8,10]



parameter_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
grid_search = GridSearchCV(estimator = classifier_rf,
                           param_grid = parameter_grid,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_
```

```python
classifier_rf_tuned=RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=10, max_features='sqrt', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=200, n_jobs=None,
            oob_score=False, random_state=0, verbose=0, warm_start=False)
```

```python
classifier_rf_tuned.fit(X_train, y_train)
```

```python
y_train_pred = classifier_rf_tuned.predict(X_train)
y_test_pred = classifier_rf_tuned.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
print('percentage accuracy on training set: ',accuracy_score(y_train, y_train_pred)*100,'%')
```

```python
print('accuracy on testing set: ',accuracy_score(y_test, y_test_pred)*100,'%' )
```

```python
accuracy_score(y, classifier_rf_tuned.predict(x))
```
